```python
import pandas
STTa=pandas.read_pickle("STTa.pkl")
STTm=pandas.read_pickle("STTm.pkl")
data_matrix=STTm.T.to_numpy()
```

```python
from spalor.util.randomized_matrix_computations import leverage_scores_QR
ls=leverage_scores_QR(data_matrix.transpose(),r=3)
ls
```

```python
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
plt.plot(ls,'o')
plt.show()
```

```python
ls=ls.transpose()
plt.plot(ls,'ob')
plt.plot(np.squeeze(cols), ls[cols], 'or')
plt.show()
```
